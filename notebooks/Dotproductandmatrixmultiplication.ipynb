```python
import torch
```

```python
a = torch.tensor([1,2,3])
b = torch.tensor([4,5,6])
```

```python
print(torch.dot(a,b))
```

```python
a = torch.tensor([1.0,2.0,3.0])
b = torch.tensor([4.0,5.0,6.0])

print(torch.dot(a,b))
```

```python
A = torch.tensor([
    [1,2],
    [3,4]
], dtype=torch.float32)

B = torch.tensor([
    [5,6],
    [7,8]

], dtype = torch.float32)

print(torch.matmul(A,B))
```
