```python
import torch
```

```python
scalar = torch.tensor(7)

vector = torch.tensor([1,2,3,4,5])

matrix = torch.tensor([[1,2,3],
                       [4,5,6]])

random_tensor = torch.rand(3,2)

print("SCALAR")
print("value:", scalar)
print("shape:", scalar.shape)
print("ndim:", scalar.ndim)
print("dtype:", scalar.dtype)
print()

print("VECTOR")
print("value:", vector)
print("shape:", vector.shape)
print("ndim:", vector.ndim)
print("dtype:", vector.dtype)
print()

print("MATRIX")
print("value:", matrix)
print("shape:", matrix.shape)
print("ndim:", matrix.ndim)
print("dtype:", matrix.dtype)
print()

print("RANDOM TENSOR")
print("value:", random_tensor)
print("shape:", random_tensor.shape)
print("ndim:", random_tensor.ndim)
print("dtype:", random_tensor.dtype)
```
