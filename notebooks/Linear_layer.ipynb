```python
import torch
```

```python
from torch import nn
```

```python
class CreditNet(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = nn.Linear(4,2)

    def forward(self, x):
        return self.linear(x)
```

```python
model = CreditNet()
print(model)
```

```python
for name, param in model.named_parameters():
    print(name)
    print(param)
    print(param.shape)
    print("-" * 40)
```

```python
x = torch.tensor([
    [50000.0, 29.0, 10000.0, 1.0],
    [70000.0, 31.0, 7000.0, 0.0]

])

output = model(x)
print(output)
print(output.shape)
```

```python
single_x = torch.tensor([[50000.0, 29.0, 10000.0, 1.0]])
print(single_x.shape)

single_output = model(single_x)
print(single_output)
print(single_output.shape)
```
